```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
```

```python
df=pd.read_csv("/content/drive/MyDrive/Salary_Data.csv")
df
```

```python
plt.scatter(df['YearsExperience'],df['Salary'])
plt.show()
```

```python
X=df[['YearsExperience']]
y=df[['Salary']]
model=LinearRegression()
model.fit(X,y)
y_pred=model.predict(X)
y_pred
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y12=model.coef_*12+model.intercept_
y12
```

```python
print(mean_squared_error(y_pred,y,squared=False))
print(r2_score(y_pred,y))
print(mean_absolute_error(y_pred,y))
print(mean_squared_error(y_pred,y,squared=True))
```

```python
df1=pd.read_csv("/content/drive/MyDrive/Melbourne_housing_FULL.csv")
df1
```

```python
num_col=df1.dtypes[df1.dtypes!='object'].index.values
cat_col=df1.dtypes[df1.dtypes=='object'].index.values
print(num_col,cat_col)
```

```python
df1[cat_col].nunique() #number of unique values
```

```python
for i in cat_col:
  if df1[i].nunique()>50:
    df1=df1.drop(i,axis=1)
```

```python
df1
```

```python
a=df1[['Rooms','Bedroom2']].dropna()
print(a)
(a['Rooms']==a['Bedroom2']).value_counts()
```

```python
df1=df1.drop(columns=['Bedroom2','Lattitude','Longtitude','Postcode'])
df1
```

```python
cols=['Distance','Bathroom','Car','Propertycount']
df1[cols]=df1[cols].fillna(0)
df1['Landsize']=df1['Landsize'].fillna(df1['Landsize'].mean())
df1['BuildingArea']=df1['BuildingArea'].fillna(df1['BuildingArea'].mean())
df1['YearBuilt']=2024-df1['YearBuilt']
```

```python
df1=df1.dropna()
```

```python
df2=df1.copy().select_dtypes(exclude=['object'])
df2
```

```python
for i in df2.columns:
  plt.scatter(df2[i],df2['Price'])
  plt.show()
```

```python
X=df2.drop('Price',axis=1)
y=df2[['Price']]
X_train,y_train,X_test,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
model=LinearRegression()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
```
