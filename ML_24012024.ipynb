```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,accuracy_score
import seaborn as sns
```

```python
df=pd.read_csv("ChurnData.csv")
df
```

```python
df=df[['tenure','age','address','income','ed','employ','equip','callcard','wireless','churn']]
df
```

```python
df.columns
```

```python
l1=df.columns
for i in l1:
  plt.scatter(df[i],df['churn'])
  plt.xlabel(i)
  plt.ylabel("churn")
  plt.figure(figsize=(2,2),dpi=50)
  plt.show()
```

```python
df1=pd.read_csv("penguins.csv")
df1
```

```python
df1.nunique()
```

```python
for i in df1.columns:
  plt.scatter(df1[i],df1['species'])
  plt.xlabel(i)
  plt.ylabel("species")
  plt.show()
```

```python
sns.heatmap(df1.corr(),annot=True)
```

```python
df2=df1.drop(['species','island','year'],axis=1)
X_train,X_test,y_train,y_test=train_test_split(df2,df1['species'],test_size=0.2)
```

```python
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
print(y_pred)
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
accuracy_score(y_test,y_pred)
```
